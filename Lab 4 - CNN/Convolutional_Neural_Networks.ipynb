## Convolutional Neural Networks (CNN)

Machine learning is a field of artificial intelligence that enables systems to learn from data and make decisions without being explicitly programmed. It involves algorithms that iteratively learn from data and improve their accuracy over time by optimizing an error metric, typically through the use of loss functions. These functions quantify the difference between the predicted outputs and the actual outputs, guiding the algorithm to minimize this error during training. By continually adjusting and improving, machine learning models can achieve remarkable accuracy in tasks ranging from simple classification to complex decision-making scenarios.

Convolutional Neural Networks (CNNs) are a specialized kind of neural networks particularly effective for analyzing visual imagery. They employ a mathematical operation known as convolution, which uses sliding window techniques to process data. This method is highly efficient in recognizing patterns and features in images due to the sparse interactions and parameter sharing of convolutional layers. These characteristics allow CNNs to capture local patterns like edges and textures with a significantly reduced amount of parameters compared to fully connected networks, making them highly efficient for tasks involving high-dimensional data like images.

### Key Components of CNN Architecture

The architecture of a Convolutional Neural Network (CNN) typically includes several key components, each playing a crucial role in the network's ability to process and interpret visual information:

- **Convolutional Layers**:
  - Apply a number of filters to the input.
  - Create feature maps that capture essential details within the data.

- **Pooling Layers**:
  - Often follow convolutional layers.
  - Reduce the spatial size of the feature maps, decreasing the computational load.
  - Extract dominant features that are invariant to small changes in the input.

- **Fully Connected Layers**:
  - Interpret the features extracted by convolutional and pooling layers.
  - Perform classification or regression tasks based on the interpreted features.

- **Softmax Layer**:
  - Typically used in the final layer if the task is classification.
  - Outputs the probabilities of the instance belonging to each class.

- **Dropout Layers**:
  - Included to prevent overfitting.
  - Randomly drop units from the neural network during the training process to improve generalization.

### Overview of Neural Network Hyperparameters

In training neural networks, various hyperparameters must be predefined to guide the training process. These include:

- **Batch Size**
  - **Definition**: Number of training examples used per iteration.
  - **Common Values**: 32, 64, 128, 256.

- **Learning Rate**
  - **Definition**: Step size at each iteration to minimize the loss function.
  - **Common Values**: Often starts small, e.g., 0.01, 0.001, and can be dynamically adjusted (lowered as we get nearer to minima, check scheduers).

- **Optimizer Type**
  - **Examples**: SGD (Stochastic Gradient Descent), Adam, RMSprop, Adagrad.
  - **Role**: Different optimizers affect training dynamics and model performance differently.

- **Beta Parameters for Adam Optimizer**
  - **Beta1 and Beta2**: Control the decay rates of moving averages of past gradients and squared gradients.
  - **Common Values**: Beta1 = 0.9, Beta2 = 0.999.

- **Epochs**
  - **Definition**: One full pass of the training dataset through the learning algorithm.
  - **Usage**: More epochs can lead to better learning but risk overfitting.

- **Momentum**
  - **Definition**: Helps accelerate SGD and dampens oscillations.
  - **Common Values**: 0.9, 0.99.

- **Regularization Parameters (L1 & L2)**
  - **Purpose**: Prevent overfitting by adding a penalty on the size of the coefficients.
  - **Parameters**: Lambda or alpha values for L1 (Lasso) and L2 (Ridge) regularization.

- **Dropout Rate**
  - **Definition**: Fraction of neurons to randomly drop during training.
  - **Common Values**: 0.2, 0.3, 0.5.

- **Number of Layers and Number of Neurons in Each Layer**
  - **Role**: Determines the architecture's depth and width, affecting its ability to capture complex patterns.

- **Error Function Type for Classification with Softmax**
  - **Definition**: Cross-entropy loss, also known as log loss.
  - **Purpose**: Measures the performance of a classification model whose output is a probability value between 0 and 1. Cross-entropy loss increases as the predicted probability diverges from the actual label (an alternative can be Categorical Hinge Loss).

- **Early Stopping**
  - **Purpose**: Halt training when performance on a validation set starts to worsen to prevent overfitting.

Each of these hyperparameters can significantly impact the effectiveness and efficiency of the neural network training process, and tuning them appropriately is crucial for achieving optimal performance.

````python
# DO NOT CHANGE THIS CELL
## Training set indices.
np.random.seed(27)  # Set random seed for reproducibility

# Randomly select indices for the test sets for each class
residential_indices = np.random.choice(np.arange(3000), size=600, replace=False)
forest_indices = np.random.choice(np.arange(3000), size=600, replace=False)
river_indices = np.random.choice(np.arange(2500), size=500, replace=False)

residential_val_indices = residential_indices[:300]
forest_val_indices = forest_indices[:300]
river_val_indices = river_indices[:250]

residential_test_indices = residential_indices[300:]
forest_test_indices = forest_indices[300:]
river_test_indices = river_indices[250:]
````

````python
def natural_sort_key(s):
    """Key function for natural sorting."""
    return [int(text) if text.isdigit() else text.lower() for text in re.split(r'(\d+)', s)]

def load_images(class_name):
    images = []
    dataset_path = '/content/dataset/EuroSATallBands/ds/images/remote_sensing/otherDatasets/sentinel_2/tif/'
    class_path = os.path.join(dataset_path, class_name)

    # Use glob to find files matching the pattern '*.tif' and sort them numerically
    filenames = sorted(glob.glob(os.path.join(class_path, '*.tif')), key=natural_sort_key)

    for filename in filenames:
        image = tifffile.imread(filename)
        images.append(image)

    return images

# Load images for each class
residential_images = load_images('Residential')
river_images = load_images('River')
forest_images = load_images('Forest')

print("Number of residential images:", len(residential_images))
print("Number of river images:", len(river_images))
print("Number of forest images:", len(forest_images))
````

````python
# Function to display images in a row
def display_images(images, title):
    fig, axs = plt.subplots(1, 10, figsize=(20, 4))
    fig.suptitle(title, fontsize=16)

    for i in range(10):
        axs[i].imshow(images[i][:, :, 3])
        axs[i].axis('off')

    plt.show()

# Display the first 5 residential images
display_images(residential_images[:10], 'Residential Images')

# Display the first 5 river images
display_images(river_images[:10], 'River Images')

# Display the first 5 forest images
display_images(forest_images[:10], 'Forest Images')
````

````python
# Get the training indices using setdiff1d
residential_train_indices = np.setdiff1d(np.arange(3000), np.concatenate([residential_val_indices, residential_test_indices]))
forest_train_indices = np.setdiff1d(np.arange(3000), np.concatenate([forest_val_indices, forest_test_indices]))
river_train_indices = np.setdiff1d(np.arange(2500), np.concatenate([river_val_indices, river_test_indices]))

# Check the lengths of the training sets
print("Residential - Train:", len(residential_train_indices))
print("Forest - Train:", len(forest_train_indices))
print("River - Train:", len(river_train_indices))
````

````python
# Split the dataset
residential_train = [residential_images[i] for i in residential_train_indices]
forest_train = [forest_images[i] for i in forest_train_indices]
river_train = [river_images[i] for i in river_train_indices]

residential_val = [residential_images[i] for i in residential_val_indices]
forest_val = [forest_images[i] for i in forest_val_indices]
river_val = [river_images[i] for i in river_val_indices]

residential_test = [residential_images[i] for i in residential_test_indices]
forest_test = [forest_images[i] for i in forest_test_indices]
river_test = [river_images[i] for i in river_test_indices]

# Check the lengths of the splits
print("Residential - Train:", len(residential_train), "Validation:", len(residential_val), "Test:", len(residential_test))
print("Forest - Train:", len(forest_train), "Validation:", len(forest_val), "Test:", len(forest_test))
print("River - Train:", len(river_train), "Validation:", len(river_val), "Test:", len(river_test))
````

````python
class EuroSATDataset(Dataset):
    def __init__(self, images, labels, transform=None):
        self.images = images
        self.labels = labels
        self.transform = transform

    def __len__(self):
        return len(self.images)

    def __getitem__(self, idx):
        image = self.images[idx]
        label = self.labels[idx]

        if self.transform:
            image = self.transform(image)

        return image, label

# Custom transform to stack the bands
class StackBands(object):
    def __init__(self, dtype=np.float32):
        self.dtype = dtype

    def __call__(self, image):
        # Convert image bands to a supported data type
        image = [band.astype(self.dtype) for band in image]

        # Stack the bands along the channel dimension
        image = np.stack(image, axis=0)
        return image

# Data loading and preprocessing
data_transform = transforms.Compose([
    StackBands(dtype=np.float32),  # Use float32 for compatibility with to_tensor
    transforms.ToTensor(),
])


# Combine the training datasets
train_images = residential_train + forest_train + river_train
train_labels = [0] * len(residential_train) + [1] * len(forest_train) + [2] * len(river_train)

# Combine the validation datasets
val_images = residential_val + forest_val + river_val
val_labels = [0] * len(residential_val) + [1] * len(forest_val) + [2] * len(river_val)

# Combine the testing datasets
test_images = residential_test + forest_test + river_test
test_labels = [0] * len(residential_test) + [1] * len(forest_test) + [2] * len(river_test)

# Create datasets and data loaders
train_dataset = EuroSATDataset(train_images, train_labels, transform=data_transform)
val_dataset = EuroSATDataset(val_images, val_labels, transform=data_transform)
test_dataset = EuroSATDataset(test_images, test_labels, transform=data_transform)

# Create data loaders
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=32, shuffle=False)
test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False)

````

````python
# Define the CNN architecture
class MultiBandCNN(torch.nn.Module):
    def __init__(self):
        super(MultiBandCNN, self).__init__()

        self.conv1 = torch.nn.Conv2d(13, 32, kernel_size=3, padding=1)  # First convolutional layer with 13 input channels, 32 output channels, 3x3 kernel size, and 1 pixel padding
        self.conv2 = torch.nn.Conv2d(32, 64, kernel_size=3, padding=1)  # Second convolutional layer with 32 input channels, 64 output channels, 3x3 kernel size, and 1 pixel padding
        self.conv3 = torch.nn.Conv2d(64, 128, kernel_size=3, padding=1) # Third convolutional layer with 64 input channels, 128 output channels, 3x3 kernel size, and 1 pixel padding
        self.pool = torch.nn.MaxPool2d(kernel_size=2, stride=2)         # Max pooling layer with a kernel size of 2x2 and a stride of 2
        self.fc1 = torch.nn.Linear(128 * 8 * 8, 512)                    # The first fully connected (linear) layer with an input size of 128*8*8 and an output size of 512
        self.fc2 = torch.nn.Linear(512, 3)                              # The second fully connected (linear) layer with an input size of 512 and an output size of 3 (for 3 output classes)

        # 128 * 8 * 8:
        # 128: This comes from the number of output channels of the last convolutional layer (conv3), which is 128. Each channel represents a different feature map resulting from the convolution operation.
        # 8 * 8: This comes from the spatial dimensions of the feature maps after passing through the max pooling layers (pool).
        #        Each max pooling operation with a kernel size of 2x2 and a stride of 2 reduces the spatial dimensions by half. Starting with an input image size of 64x64 (typical for EuroSAT dataset), the first pooling layer reduces it to 32x32, then 16x16, and finally 8x8.
        #512: This is an arbitrary choice for the size of the hidden layer in the network. The fully connected layer fc1 takes the flattened output of the last convolutional layer and maps it to a hidden representation of size 512. This hidden layer acts as a bottleneck, compressing the information learned from the convolutional layers before mapping it to the output layer.

    def forward(self, x):
        x = self.pool(torch.relu(self.conv1(x)))          # Apply the first convolutional layer followed by ReLU activation and max pooling
        x = self.pool(torch.relu(self.conv2(x)))          # Apply the second convolutional layer followed by ReLU activation and max pooling
        x = self.pool(torch.relu(self.conv3(x)))          # Apply the third convolutional layer followed by ReLU activation and max pooling
        x = x.view(-1, 128 * 8 * 8)                       # Flatten the output for the fully connected layers
        x = torch.relu(self.fc1(x))                       # Apply the first fully connected layer followed by ReLU activation
        x = self.fc2(x)                                   # Apply the second fully connected layer
        return x

# Create an instance of the MultiBandCNN model
model = MultiBandCNN()

# Set hyperparameters
learning_rate = 0.001
num_epochs = 10

# Define the loss function (CrossEntropyLoss) for classification
criterion = torch.nn.CrossEntropyLoss()
# Define the optimizer (Adam) for updating the model parameters
optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)



'''
Justification of architecture selection:
- Convolutional Layers: The network starts with three convolutional layers (conv1, conv2, conv3) with increasing depth.
                         These layers are essential for learning spatial hierarchies of features in the input images, which is important for distinguishing between different land cover types.

- Pooling Layers: After each convolutional layer, a max pooling layer (pool) is applied.
                  This helps in reducing the spatial dimensions of the feature maps while retaining the most important information, thus aiding in computational efficiency and reducing overfitting.

- Fully Connected Layers: The flattened output from the last convolutional layer is passed through two fully connected layers (fc1, fc2) for classification.
                           These layers help in learning non-linear combinations of features learned by the convolutional layers, allowing the model to make predictions based on these higher-level features.

- ReLU Activation: Rectified Linear Unit (ReLU) activation functions are used after each convolutional and fully connected layer.
                   ReLU helps introduce non-linearity to the model, allowing it to learn complex patterns in the data.
'''
````

````python
# Calculate F1 score for the best model
def calculate_f1_score(y_true, y_pred):
    classes = np.unique(y_true)
    f1_scores = []
    for c in classes:
        tp = np.sum((y_true == c) & (y_pred == c))
        fp = np.sum((y_true != c) & (y_pred == c))
        fn = np.sum((y_true == c) & (y_pred != c))
        precision = tp / (tp + fp + 1e-9)
        recall = tp / (tp + fn + 1e-9)
        f1 = 2 * (precision * recall) / (precision + recall + 1e-9)
        f1_scores.append(f1)
    macro_f1 = np.mean(f1_scores)
    return macro_f1
````

````python
# Set the ranges of learning rates and num_epochs to try
learning_rates = [0.001, 0.01, 0.1]
num_epochs_list = [5, 10, 30]

best_model = None
best_accuracy = 0

for learning_rate in learning_rates:
    for num_epochs in num_epochs_list:
        # Define your model, optimizer, and criterion
        print(num_epochs)
        print(learning_rate)
        model = MultiBandCNN()
        optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)
        criterion = torch.nn.CrossEntropyLoss()

        # Training loop
        for epoch in range(num_epochs):
            model.train()
            for images, labels in train_loader:
                optimizer.zero_grad()
                outputs = model(images)
                loss = criterion(outputs, labels)
                loss.backward()
                optimizer.step()


        # Training accuracy
        model.eval()
        y_true_train = []
        y_pred_train = []
        with torch.no_grad():
            for images, labels in train_loader:
                outputs = model(images)
                _, predicted = torch.max(outputs, 1)
                y_true_train.extend(labels.numpy())
                y_pred_train.extend(predicted.numpy())

        train_accuracy = sum([1 if yt == yp else 0 for yt, yp in zip(y_true_train, y_pred_train)]) / len(y_true_train)
        print(f"Training Accuracy: {train_accuracy}")

        # Validation accuracy
        model.eval()
        y_true_val = []
        y_pred_val = []
        with torch.no_grad():
            for images, labels in val_loader:
                outputs = model(images)
                _, predicted = torch.max(outputs, 1)
                y_true_val.extend(labels.numpy())
                y_pred_val.extend(predicted.numpy())

        val_accuracy = sum([1 if yt == yp else 0 for yt, yp in zip(y_true_val, y_pred_val)]) / len(y_true_val)
        print(f"Validation Accuracy: {val_accuracy}")


        # Test accuracy
        model.eval()
        y_true_test = []
        y_pred_test = []
        with torch.no_grad():
            for images, labels in test_loader:
                outputs = model(images)
                _, predicted = torch.max(outputs, 1)
                y_true_test.extend(labels.numpy())
                y_pred_test.extend(predicted.numpy())

        test_accuracy = sum([1 if yt == yp else 0 for yt, yp in zip(y_true_test, y_pred_test)]) / len(y_true_test)
        print(f"Test Accuracy: {test_accuracy}")
        f1 = calculate_f1_score(y_true_test, y_pred_test)
        print(f"Average F1 Score: {f1}")

        # If this model has the best accuracy, store it
        if test_accuracy > best_accuracy:
            best_accuracy = test_accuracy
            best_model = model
            best_confusion_matrix = confusion_matrix(y_true_test, y_pred_test)

# Print the best accuracy
print(f"Best Test Accuracy: {best_accuracy}")

# Visualize confusion matrix using seaborn heatmap for the best model
plt.figure(figsize=(8, 6))
sns.heatmap(best_confusion_matrix, annot=True, fmt='d', cmap='Blues', cbar=False,
            xticklabels=['Residential', 'River', 'Forest'],
            yticklabels=['Residential', 'River', 'Forest'])
plt.xlabel('Predicted labels')
plt.ylabel('True labels')
plt.title('Confusion Matrix for Best Model')
plt.show()
````
